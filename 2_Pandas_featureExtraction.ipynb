```python
df = pd.read_csv("anime-checkpoint.csv")
df
```

```python
def epi(txt):
    check = False
    data = ""
    for i in txt:
        if i == ")":
            check = False
            return data
        if check == True:
            data += i
        if i == "(":
            check = True
            
```

```python
df["Episodes"] = df["Title"].apply(epi)
df
```

```python
df["Episodes"] = df["Episodes"].str.replace(" eps","")
```

```python
df["Episodes"] = df["Episodes"].astype(int)
df
```

```python
#Make a new column for episode count
#make a new column for time stamp
df.loc[1]["Title"]
```

```python
def extraction_timestamp(txt):
    check = False
    data = ""
    for i in range(len(txt)):
        if txt[i] == ")":
            check = True
            for j in range(i+1,i+20):
                data += txt[j]
            return data
```

```python
df["Time"] = df["Title"].apply(extraction_timestamp)
```
