```python
# Excersice
#columns : [Age,math,science]
data = np.array([
    [18,85,78],
    [19,92,88],
    [17,76,95],
    [18,65,70],
    [20,90,85]
])
```

```python
#shape of matrix
data.shape
```

```python
#find avg age of all student
np.mean(data[:,0])
```

```python
#extarct math marks of all students
data[:,1]
```

```python
#find highest science marks
data[:,2].max()
```

```python
# get details of student who scored more than 90 in Math
data[data[:,1]>90]
```

```python
#increase math marks of all students by 5
data[:,1] +=5
data
```

```python
# Find how many students younger then 19
data[data[:,0] <19]
```

```python
# calculate the avg marks in each subject [column-wise mean].
np.mean(data[:,1:],axis=0)
```

```python
# get data of students who scored at least 80 in both subject.
data[(data[:,1] >= 80 ) & (data[:,2] >= 80)] 
```

```python
#replace all Science marks < 75 with 0
data[:,2][data[:,2] < 75] = 0
data
```
